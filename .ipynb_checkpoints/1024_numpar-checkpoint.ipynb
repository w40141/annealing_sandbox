```python
from IsingLib import Model, Bulletforce, Reduce
import numpy as np
import random
import collections
import json
```

```python
def fib(m, n):
    a = b = m
    for _ in range(n):
        yield a
        a, b = b, a + b
```

```python
spins = 10
vartype = 'SPIN'
bit = 10
val = 2 ** (bit - 1)
mat = np.random.randint(-val, val+1, (spins, spins))
# li = np.array(list(fib(1, spins)))
# mat = np.array([[0 if i == j else li[i] * li[j] for i in range(spins)] for j in range(spins)]) * 2
# const = sum([i ** 2 for i in li])
# model = Model(vartype, h=(mat, const))
model = Model(vartype, h=mat)
```

```python
model.compile()
```

```python
ising_model, const = model.to_ising()
bf = Bulletforce(ising_model, const, vartype)
```

```python
bf.draw_eng(show=0)
eng, original_optimal_state = bf.run()
```

```python
re = Reduce(ising_model)
scr_bit = re.scr_bit
mode = 'normarlize'
```

```python
results = {}
for i in range(scr_bit, 1, -1):
    bit_result = {}
    re_mat, tmp_c = re.function(mode, i)
    re_bf = Bulletforce(re_mat, tmp_c, vartype)
    re_bf.draw_eng(name='numper_'+str(spins)+'_'+mode+'_'+str(scr_bit)+'_to_'+str(i), show=0)
    eng, re_optimal_state = re_bf.run()
    ans = [sum(state * li) for state in re_optimal_state]
    c = collections.Counter(ans)
    optimal_num = len(ans)
    optimal = c[0]
    sub_optimal = c[-2] + c[2]
    others = optimal_num - optimal - sub_optimal
    bit_result.update(optimal_num=optimal_num, optimal=optimal, sub_optimal=sub_optimal, others=others)
    results[i]=bit_result
```

```python
with open('numpar_'+str(spins)+'_'+mode+'.json', 'w') as f:
    json.dump(results, f, ensure_ascii=False)
```

```python
qubo_model, const = model.to_qubo()
re = Reduce(qubo_model)
scr_bit = re.scr_bit
mode = 'gradate'
results = {}
for i in range(scr_bit, 1, -1):
    bit_result = {}
    re_mat, tmp_c = re.function(mode, i)
    re_bf = Bulletforce(re_mat, tmp_c, 'qubo')
    re_bf.draw_eng(name='qubo_numper_'+str(spins)+'_'+mode+'_'+str(scr_bit)+'_to_'+str(i),)
    eng, re_optimal_state = re_bf.run()
    ans = [sum((2 * np.array(state) - 1) * li) for state in re_optimal_state]
    c = collections.Counter(ans)
    optimal_num = len(ans)
    optimal = c[0]
    sub_optimal = c[-2] + c[2]
    others = optimal_num - optimal - sub_optimal
    bit_result.update(optimal_num=optimal_num, optimal=optimal, sub_optimal=sub_optimal, others=others)
    results[i]=bit_result
print(results)
```

```python
with open('qubo_numpar_'+str(spins)+'_'+mode+'.json', 'w') as f:
    json.dump(results, f, ensure_ascii=False)
```
