```python
from IsingLib import Model, DaSampler
import numpy as np
```

```python
num = 20
spin = 5000
# mat = np.random.randint(- num ** 2, num ** 2, (spin, spin))
num_li = [20 + i for i in range(num)]
# num_li = [4, 2, 7, 1]
mat = [[i * j for i in num_li] for j in num_li]
const = sum(np.diag(mat))
mat = np.triu(mat, k=1)
model = Model('i', mat*2, const, )
```

```python
da_model = model.to_da()
print(da_model)
```

```python
da = DaSampler(**da_model, api_key='../../../da_key_2.json', version=2)
```

```python
job_id = da.sample('da')
```

```python
da.get_job(job_id)
```

```python
da.get_energy(job_id)
```

```python
for i, v in da.jobs.items():
    for j, k in v.items():
        for s, t in k.items():
            print(s, t)
```
